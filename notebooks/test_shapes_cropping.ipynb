```python
from histpat_toolkit.dzi_file import DZIFile
from histpat_toolkit.image_pyramid import DZIPyramid
from PIL import Image
import histpat_toolkit.geom as geom
import matplotlib.pyplot as plt
import cv2
import numpy as np

piramid = DZIPyramid(
    DZIFile("https://cancercenter-data.s3.sbg.io.cloud.ovh.net/b04344d8-efeb-4638-9458-870b4af4cf9f/0.dzi"))


star = geom.Polygon([
                        geom.Point(40, 140),
                        geom.Point(30, 180),
                        geom.Point(0, 180),
                        geom.Point(25, 200),
                        geom.Point(10, 240),
                        geom.Point(40, 220),
                        geom.Point(70, 240),
                        geom.Point(55, 200),
                        geom.Point(80, 180),
                        geom.Point(50, 180),
                    ])
img = piramid.full_image(scale=0.0625 / 2)
cv2.fillPoly(img, [np.array(star.points, dtype=np.int32)], (255, 0, 0))
display(Image.fromarray(img))


square = geom.Polygon([
                        geom.Point(50, 50),
                        geom.Point(150, 50),
                        geom.Point(150, 150),
                        geom.Point(50, 150),
                    ])

img = piramid.full_image(scale=0.0625 / 2)
cv2.fillPoly(img, [np.array(square.points, dtype=np.int32)], (255, 0, 0))
display(Image.fromarray(img))


diamond = geom.Polygon([
                        geom.Point(60, 30),
                        geom.Point(90, 60),
                        geom.Point(60, 90),
                        geom.Point(30, 60),
                    ])
img = piramid.full_image(scale=0.0625 / 2)
cv2.fillPoly(img, [np.array(diamond.points, dtype=np.int32)], (255, 0, 0))
display(Image.fromarray(img))


rotated_rectangle = geom.Rectangle.with_center(100, 100, 100, 100, 20)
img = piramid.full_image(scale=0.0625 / 2)
cv2.fillPoly(img, [np.array(rotated_rectangle.points(), dtype=np.int32)], (255, 0, 0))
display(Image.fromarray(img))


circle = geom.Circle(geom.Point(60, 60), 50)
img = piramid.full_image(scale=0.0625 / 2)
cv2.circle(img, (circle.center.x, circle.center.y), circle.radius, (255, 0, 0), -1)
display(Image.fromarray(img))


ellipse = geom.Ellipse(geom.Point(60, 60), (50, 30))
img = piramid.full_image(scale=0.0625 / 2)
cv2.ellipse(img, (ellipse.center.x, ellipse.center.y), (int(ellipse.axes[0]), int(ellipse.axes[1])), 0, 0, 360, (255, 0, 0), -1)
display(Image.fromarray(img))


cropped_star = piramid.crop_shape(star, fill_value=0, scale=0.0625 / 2)
cropped_square = piramid.crop_shape(square, fill_value=0, scale=0.0625 / 2)
cropped_diamond = piramid.crop_shape(diamond, fill_value=0, scale=0.0625 / 2)
cropped_rotated_rectangle = piramid.crop_shape(rotated_rectangle, fill_value=0, scale=0.0625 / 2)
cropped_circle = piramid.crop_shape(circle, fill_value=0, scale=0.0625 / 2)
cropped_ellipse = piramid.crop_shape(ellipse, fill_value=0, scale=0.0625 / 2)


display(Image.fromarray(cropped_star))
display(Image.fromarray(cropped_square))
display(Image.fromarray(cropped_diamond))
display(Image.fromarray(cropped_rotated_rectangle))
display(Image.fromarray(cropped_circle))
display(Image.fromarray(cropped_ellipse))
```
